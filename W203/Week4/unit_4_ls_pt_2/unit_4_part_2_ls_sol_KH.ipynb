$$f_X(x)=\begin{cases} x, & 0 \leq x < 1 \\ 2 - x, & 1 \leq x < 2 \\ 0, & \text{otherwise} \end{cases}$$

**3.2** Find the cumulative distribution function of $X$ and plot it.

>In a cumulative probability distribution, $F(x)$, is the area under f (the probability distribution function) from $-\infty$ to $x$
>
>$$F(x)=P(X \leq x) = \int_{y=-\infty}^x{f(y)dy}$$
>
>The probability distribution function for X as described above, $f_X(x)$, was given as 
>
>$$\text{} f_X(x)=\begin{cases} x, & 0 \leq x < 1 \\ 2 - x, & 1 \leq x < 2 \\ 0, & \text{otherwise} \end{cases}$$

>To compute the cumulative probability function of X, $F_X(x)$, we need to integrate piece from $f_X(x)$ individually
>
>Step 1: Get the formula for the CDF from (0$\rightarrow$1)
$$\text{For } x \in (0,1) \quad F_X(x) = \int_{0}^{x}{f_X(x)}dx = \int_{0}^{x}u \cdot du$$

>Get the integral
$$F_X(x) = \left[\frac{ u^2}{2} \right]_0^x$$




>Evaluate over the interval
$$F_X(x)= \left[\frac {x^2}{2} - \frac {0^2}{2} \right]$$



>Simplify
$$\text{For } x \in (0,1) \quad F_X(x) =  \left[\frac {x^2}{2}\right]$$

>Step 2: Get the formula for the CDF from (0$\rightarrow$2). Note (0$\rightarrow$2) includes (0$\rightarrow$1) in the cummulation
>
>$$\text{For } x \in (0,2) \quad F_X(x) = \int_{0}^{1}{f_X(x)}dx + \int_{1}^x{f_X(x)}dx = \int_{0}^{1}u \cdot du + \int_{1}^{x}(2-u)\cdot du$$

>Get the integral
$$F_X(x) = \left[\frac {u^2}{2} \right]_0^1 + \left[2u-\frac {u^2}{2}\right]_1^x$$  



>Evaluate over the interval
>$$F_X(x) = \left[\frac {1^2}{2} - \frac {0^2}{2} \right] + \left[(2x-\frac {x^2}{2})-(2-\frac {1^2}{2})\right]$$
>
>$$F_X(x) = \left[\frac {1}{2}\right]+ \left[(2x-\frac {1}{2}x^2)-1\frac{1}{2} \right]$$

>Simplify
$$\text{For } x \in (0,2) \quad F_X(x) = 2x-\frac {1}{2}x^2-1$$

>Assemble the full CDF from the individual CDF components for each of the ranges
>
> $$F_X(x)=\begin{cases} 0 &;& x < 0 \\
\frac{x^2}{2} &;& 0 \leq x < 1 \\
2x - \frac{1}{2}x^2 -1 &;& 1 \leq x < 2 \\
1 &;&  2 < x \end{cases}$$

```r
x1 <- seq(-2, 0, 0.1)
x2 <- seq(0,1, 0.1)
x3 <- seq(1,2, 0.1)
x4 <- seq(2,4, 0.1)

y1 <- 0*x1 # rep(0, length(x1))
y2 <- x2**2/2
y3 <- 2*x3 - (x3**2)/2 - 1
y4 <- 1*x4/x4

plot (c(-2,4), c(0,1), type="n", xlab="x", ylab="y")
lines(x1, y1, col="red", lwd=2)
lines(x2, y2, col="blue", lwd=2)
lines(x3, y3, col="green", lwd=2)
lines(x4, y4, col="yellow", lwd=2)

leg = legend("topleft",
      inset=.05,
      cex = 1, 
       title ="Legend",
      legend = c("F_x(X)=x^2/2", "F_x(X)=2x-1/2x^2-1"),
      horiz=TRUE,
      lty=c(1,1),  
      lwd=c(2,2),
      col=c("red","blue"),
      bg="grey96",
      text.font=3 
      )
```

0, & x < 0 \\

**3.3** Compute the expectation of $X$.

>The expectation of X is defined as
>
>$$E(X)=\int_{-\infty}^{\infty} x \cdot f_X(x)dx$$
>
>And the probability distribution function for X as described above, $f_X(x)$, was described as 
>
>$$f_X(x)=\begin{cases} x, & 0 \leq x < 1 \\ 2 - x, & 1 \leq x < 2 \\ 0, & \text{otherwise} \end{cases}$$

>Note that again we need to integrate $x f_X(x)$ across each range to calculate E(X).
>
>
>$$E(X) = \int_{0}^{1}x \cdot {f_X(x)}dx + \int_{1}^2 x \cdot {f_X(x)}dx = \int_{0}^{1}u^2 du + \int_{1}^{2}u(2-u) du$$


>Get the integral
>$$E(X) = \left[\frac{ u^3}{3} \right]_0^1 + \left[u^2 - \frac{u^3}{3} \right]_1^2$$

>Evaluate over the intervals
>$$E(X) = \left[\frac {1^3}{3} - \frac {0^3}{3} \right] + \left[(2^2 - \frac {2^3}{3}) - (1^2 - \frac {1^3}{3}) \right]$$

>Simplify
>$$E(X) =  \left[\frac {1}{3}\right] + \left[(4 - \frac{8}{3}) - (1 - \frac {1}{3})\right] $$
>
>$$ =  \frac {1}{3} + \frac{4}{3} - \frac{2}{3} $$
>
>$$ = 1$$

<h3> 5 Modeling an American Delicacy </h3>

Suppose that the number of calories in a Twinkie is a normally-distributed random variable.Suppose further that 50% of Twinkies have over 100 calories, and 68.2% of Twinkies have calories in the interval from 85 calories to 115 calories. 

**5.1**  What fraction of Twinkies have more than 140 calories? 

> We know three things about the distribution of calories in Twinkies
1. The distribution is normal, and thus characterized by its mean and variance
2. The mean is is $100$
    1. Why? Because the median (50\% percentile) of the normal distribution is equal to its mean
3. The standard deviation is (approximately) $15$
    1. Why? Because about $68.2\%$ of the probability mass of the normal distribution lies one standard deviation away from the mean.
    2. This is known as the "68–95–99.7 rule":
        1. About $68\%$ of the distribution lies within one standard deviation from the mean
        2. About $95\%$ of the distribution lies within two standard deviations from the mean
        3. About $99.7\%$ of the distribution lies within three standard deviations from the mean

>If you don't remember the rule, you could also compute it using R.  The problem talks about a region with $68.2\%$ of the probability mass around the mean.  Subtracting this from $1$, we know that $1 - 68.2\%$ of the probability mass is in the left and right tails outside this interval.  Since the interval is symmetric about the mean, we just divide this by 2 to get the mass in the left tail.  Now, we can pass the result into R's qnorm function to find how many standard deviations away this point is.

```r
qnorm(0.5*(1-.682))
```

>The result is approximately 1, as we expect.  Now, we can use pnorm to find out the fraction of Twinkies are over 140 calories.  Note that we have to subtract from 1 since pnorm gives us the probability to the left of a value, and we want the probability to the right.

```r
1-pnorm(140, mean=100, sd=15)
```
